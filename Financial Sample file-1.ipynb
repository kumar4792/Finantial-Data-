```python
import tabula
```

```python
pdf_path = "D:\pdf document\APPLE.pdf"
```

 ### page no-37 page contain only table 

```python
dfs = tabula.read_pdf(pdf_path, pages=37, guess=True, stream=True, lattice= False ,relative_area= False ,multiple_tables=False,
                     pandas_options={"header": [0,1]},)
dfs[0]
```

```python
dfs[0].to_csv("D:\pdf document\page37.csv", index =False) # Download file to local decice in csv format
```

### Page no - 42 Page contains paragraph and table

```python
dfs42 = tabula.read_pdf(pdf_path, pages=42, guess=False,stream=True, lattice= False ,relative_area= True, multiple_tables=False,
                      pandas_options={"header": [0]})
dfs42[0]
```

 ### From Page no - 42 extracting only table

```python
dfs42_1 = tabula.read_pdf(pdf_path, pages=42, guess=False, stream=True, lattice= False ,relative_area= True,area=(68,0,90, 90),
                       multiple_tables=False)
                      
dfs42_1[0]
```

```python
dfs42_1[0].to_csv("D:\pdf document\page42.csv", index =False)
```
